Import libraries

```python
import numpy as np
import pandas as pd
```

Import Dataset

```python
raw_data = pd.read_excel('Annual_Public_Finance (1).xlsx')
```

Extract relevant data

```python
raw_data = raw_data.iloc[8:17,2:]
```

Set correct header

```python
raw_data.columns = raw_data.loc[8]
raw_data = raw_data.reset_index()
```

Cleaning Dataset

```python
raw_data = raw_data.loc[1:,'Items':]
```

```python
raw_data
```

Dataset Transpose

```python
raw_data = raw_data.T
```

Set header name

```python
raw_data.columns = raw_data.loc['Items']
```

Drop irrelevant row

```python
raw_data = raw_data.drop('Items', axis=0)
```

Reset Index

```python
raw_data = raw_data.reset_index()
```

```python
raw_data.columns
```

```python
raw_data.columns = ['Year','Country','Public Debt (Internal)', 'Public Debt (External)',
                  'Tax revenue',             'Direct tax',
               'Total receipts',      'Total expenditure',
         'Gross Fiscal Deficit']
```

Manage null value

```python
raw_data = raw_data.replace(to_replace=['--','..'], value=np.NaN)
#raw_data['Public Debt (External)'].astype(float)
```

Set datatime function to year column

```python
raw_data['Year'] = pd.to_datetime(raw_data['Year']).dt.year
final_data = raw_data.round(2)
final_data
```

Download the clean dataset

```python
final_data.info()
```

```python
final_data.to_csv('Public Finance.csv', index=False)
```
